```python

numbers = (20, 10, 16, 19, 25, 1, 276, 188)


print("Odd numbers in the list:")
for num in numbers:
    if num % 2 != 0:  
        print(num,"is Odd")
```

```python
numbers=(20, 10, 16, 19, 25, 1, 276, 188)

print("Even numbers in the list:")
for num in numbers:
    if num % 2!= 1:
        print(num,"is Even")
```
